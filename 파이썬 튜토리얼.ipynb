4. 그래프 좌표 입력 및 출력

```python
# 좌표 입력, 그래프 출력 모듈
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 그래프 좌표 입력
L = np.array([[2,2],[3,5],[5,3],[6,1]])

# 그래프 칸수를 입력
G = 6

# 그래프 그리기
x,y = L.T
C = len(L)
i = 0
while i < C:
  plt.scatter(x[i], y[i], c='skyblue', s=500)
  plt.text(x[i], y[i], i, ha='center', va='center')
  i = i+1
plt.xlim(0, G+1)
plt.ylim(0, G+1)
plt.grid(True)
plt.show()
```
