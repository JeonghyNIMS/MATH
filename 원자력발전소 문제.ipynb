```python
import numpy as np
import matplotlib.pyplot as plt

# 그래프 좌표를 순서대로 입력
L = np.array([[0,0],[1,4],[0,3],[1,2],[3,5],[3,0],[3,2],[4,4],[5,1]])

# 전체 칸수를 입력
G = 6

# 기본 포인트 수
P = 4
TP = P*2 + 1


print(L)

# 그래프 그리기
x,y = L.T
C = len(L)

for i, point in enumerate(L):
    if i == 0:
        color = 'red'
    elif i % 2 == 0:
        color = 'skyblue'
    else:
        color = 'yellow'
    plt.scatter(point[0], point[1], color=color, s=300)
    plt.text(point[0], point[1], str(i), ha='center', va='center')

plt.xlim(-0.5, G-0.5)
plt.ylim(-0.5, G-0.5)
plt.grid(True)
plt.show()

```

```python
# 포인트 간 거리
i = 0
j = 0
M = np.zeros((C,C), dtype = int)
while i < C:
  while j < C:
    M[i,j] = abs(L[i][0]-L[j][0])+abs(L[i][1]-L[j][1])
    j=j+1
  j = 0
  i=i+1

# 경로
N = [0]
i=1
while i<C:
  N.append(i)
  i=i+1

# print(N)

visit = [0] * C
arr = [0] * C
arr_list = []
def permutaion(level):
    if level >= C:
        arr_list.append(arr[:])
        return
    else:
        for i in range(C):
            if visit[i]: continue
            visit[i] = 1
            arr[level] = N[i]
            permutaion(level + 1)
            visit[i] = 0
permutaion(0)
# print(arr_list)
# print(len(arr_list))

arr_list_2 = []
i=0
while i < len(arr_list):
    if arr_list[i][0] ==0:
        if arr_list[i][1] %2 !=0:
            if arr_list[i][2] %2 ==0:
                if arr_list[i][3] %2 !=0:
                    if arr_list[i][4] %2 ==0:
                        if arr_list[i][5] %2 !=0:
                            if arr_list[i][6] %2 ==0:
                                if arr_list[i][7] %2 !=0:
                                    if arr_list[i][8] %2 ==0:
                                        arr_list_2.append(arr_list[i])
    i=i+1

# print(arr_list_2)
# print(len(arr_list_2))

# 경로 거리
arr_result=[]
i=0
while i<len(arr_list_2):
  j=0
  result=0
  while j<C-1:
    result = result+M[arr_list_2[i][j],arr_list_2[i][j+1]]
    j=j+1
  arr_result.append(result)
  i=i+1

# 최소이동거리
R=min(arr_result)
print(f"최소이동거리 = {R}")
# print(len(arr_result))

# 최소이동경로 수
RI=[]
for i in range(len(arr_result)):
  if arr_result[i]==min(arr_result):
    RI.append(i)
print(f"최소이동경로는 {int(len(RI))} 가지")

# 최소이동경로
for i in range(len(RI)):
  print(arr_list_2[RI[i]])
```
